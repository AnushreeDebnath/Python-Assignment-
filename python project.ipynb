```python
# import numpy as np

# Function to calculate least squares deviation
def least_squares(y_train, y_ideal):
    return np.sum((y_train - y_ideal) ** 2)

# Find best ideal function for each training dataset
best_ideal_mappings = {}
for train_col in train_data.columns[1:]:  # Skipping 'x' column
    min_deviation = float('inf')
    best_function = None
    
    for ideal_col in ideal_data.columns[1:]:  # Skipping 'x' column
        deviation = least_squares(train_data[train_col], ideal_data[ideal_col])
        if deviation < min_deviation:
            min_deviation = deviation
            best_function = ideal_col
    
    best_ideal_mappings[train_col] = (best_function, min_deviation)

print("Best Ideal Functions Mapping:", best_ideal_mappings)
```

```python
def map_test_data(test_data, ideal_data, best_ideal_mappings, threshold_factor=np.sqrt(2)):
    results = []
    
    for _, test_point in test_data.iterrows():
        x_val, y_test = test_point['x'], test_point['y']
        
        for train_col, (ideal_col, max_deviation) in best_ideal_mappings.items():
            # Find the corresponding y-value in the ideal function
            y_ideal = ideal_data.loc[ideal_data['x'] == x_val, ideal_col].values[0]
            deviation = abs(y_test - y_ideal)
            
            # Check if deviation is within the threshold
            if deviation <= threshold_factor * max_deviation:
                results.append({'x': x_val, 'y_test': y_test, 'ideal_function': ideal_col, 'deviation': deviation})
                
    return pd.DataFrame(results)

# Run mapping
mapped_results = map_test_data(test_data, ideal_data, best_ideal_mappings)
print(mapped_results.head())
```

```python
# import matplotlib.pyplot as plt

# Plot data for each training function and its ideal match
for train_col, (ideal_col, _) in best_ideal_mappings.items():
    plt.figure(figsize=(10, 6))
    plt.scatter(train_data['x'], train_data[train_col], label=f'{train_col} (Training Data)', color='blue')
    plt.plot(ideal_data['x'], ideal_data[ideal_col], label=f'{ideal_col} (Ideal Function)', color='yellow')
    
    # Plot mapped test points
    test_points = mapped_results[mapped_results['ideal_function'] == ideal_col]
    plt.scatter(test_points['x'], test_points['y_test'], label='Mapped Test Data', color='red')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f"{train_col} vs {ideal_col}")
    plt.legend()
    plt.show()
```
